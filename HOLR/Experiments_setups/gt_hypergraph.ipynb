```python
import graph_tool.all as gt
```

```python
# load and plot dataset
g = gt.collection.data["polbooks"]
pos = gt.sfdp_layout(g)
gt.graph_draw(g, pos=pos, bg_color="white")
```

```python
# prepare state and run MCMC
state = gt.CliqueState(g)
state.mcmc_sweep(niter=10000);
```

```python
# store the "true" cliques
cliques = []
for v in state.f.vertices():
    if state.is_fac[v]:
        continue
    print(state.c[v], state.x[v])
    if state.x[v]:
        cliques.append(state.c[v])
```
